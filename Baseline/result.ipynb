```python
def compute_map(relevance_sequence):
    precision_list = list()
    relevance_cnt = 0
    for i in range(len(relevance_sequence)):
        if relevance_sequence[i] == 0:
            pass
        else:
            relevance_cnt += 1
            precision_list.append(relevance_cnt / (i + 1))
    # print(precision_list)
    return sum(precision_list) / len(relevance_sequence)

for k, v in results_mark.items():
    MAP = compute_map(v)
    print(f"{k} - MAP={MAP:.4f}")
```
