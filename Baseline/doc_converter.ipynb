```python
import pandas as pd
import json
```

```python
document_df = pd.read_csv("joined.csv")
document_df.head(5)
```

```python
document_converted = json.loads(document_df.T.to_json())
document_simplified = [
    {
        "id": x["docid"],
        "contents": x["content"]
        } for x in document_converted.values()
]
with open("documents/documents.json", "w") as f:
    json.dump(document_simplified, f, indent=2)
```

```python
from pyserini.search import SimpleSearcher

searcher = SimpleSearcher('documents')
hits = searcher.search('documents')

for i in range(len(hits)):
    print(f'{i+1:2} {hits[i].docid:4} {hits[i].score:.5f}')
```

```python
document_df = pd.read_csv("./documents_android/documents_android.csv")
document_df.head()
```

```python
document_converted = json.loads(document_df.T.to_json())
document_simplified = [
    {
        "id": x["DocumentId"],
        "contents": x["Document Description"].replace("> **Possible Duplicate:** >", "") 
        } for x in document_converted.values()
]
with open("./documents_android/documents_android.json", "w") as f:
    json.dump(document_simplified, f, indent=2)
```
